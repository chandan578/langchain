## How to split character - character text splitter

The Character Text Splitter is a simple yet effective tool for breaking down large documents into smaller, manageable chunks based on character count. This is particularly useful for processing and analyzing text data in applications such as natural language processing, information retrieval, and machine learning. It works by specifying a maximum character limit for each chunk, ensuring that the resulting pieces of text are of a manageable size while preserving the context and meaning of the original document.


```python
from langchain_community.document_loaders import TextLoader

text_load = TextLoader('speech.txt').load()
text_load
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(separator='\n\n', chunk_size=100, chunk_overlap=20)
text_splitter.split_documents(text_load)
```

```python
speech = ""

with open('speech.txt', 'r') as f:
    speech = f.read()

text_splitter = CharacterTextSplitter(chunk_size=100, chunk_overlap=20)
text_doc = text_splitter.create_documents([speech])
```

```python
print(text_doc[0])
print(text_doc[1])
print(text_doc[2])
```
