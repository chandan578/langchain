#### Getting started With Langchain And Hugging Face

In this quickstart we will see how to :
- Set up a simple environment to use Langchain with Hugging Face models.
- Create a simple application that uses a Hugging Face model through Langchain.

```python
import os   
from dotenv import load_dotenv

load_dotenv()

os.environ['HF_TOKEN'] = os.getenv('HF_TOKEN')
os.environ['LANGCHAIN_API_KEY'] = os.getenv('LANGCHAIN_API_KEY')
os.environ['LANGCHAIN_TRACING_V2'] = 'true'
os.environ['LANGCHAIN_PROJECT'] = os.getenv('LANGCHAIN_PROJECT')
```

```python
from langchain_huggingface import ChatHuggingFace
llm = ChatHuggingFace(
    repo_id="HuggingFaceH4/zephyr-7b-beta",
    task="text-generation"
)

llm
```
