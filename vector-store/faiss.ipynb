## FAISS (Facebook AI Similarity Search)
FAISS is a library developed by Facebook AI Research that is designed for efficient similarity search and clustering of dense vectors. It is particularly useful for applications such as nearest neighbor search in high-dimensional spaces, which is common in machine learning and information retrieval tasks.
### Key Features:
- **High Performance**: FAISS is optimized for speed and can handle large datasets efficiently. It supports both CPU and GPU computations, allowing for faster processing times.
- **Scalability**: It can manage datasets that do not fit into memory by using techniques like indexing and quantization.
- **Multiple Index Types**: FAISS provides various indexing methods, including flat (brute-force), inverted file, and product quantization, allowing users to choose the best method for their specific use case.
- **Dimensionality Reduction**: It includes tools for reducing the dimensionality of vectors, which can improve search performance.
- **Clustering**: FAISS also offers clustering algorithms, such as k-means, to group similar vectors together.

```python
from langchain_community.document_loaders import TextLoader
from langchain_community.vectorstores import FAISS
from langchain_community.embeddings import OllamaEmbeddings
from langchain_text_splitters import CharacterTextSplitter
```

```python
documents = TextLoader('speech.txt').load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=30)
docs = text_splitter.split_documents(documents=documents)
```

```python
docs
```

```python
## embedding
embedding = OllamaEmbeddings(model='gemma:2b')
db = FAISS.from_documents(docs, embedding)
db
```

```python
## querying
query = 'How does the speaker describe the desired outcome of the war ??'
docs = db.similarity_search(query)
```

```python
docs
```

```python
docs[0].page_content
```

### As a Retriever:
In the context of information retrieval, FAISS can be used as a retriever to efficiently find the most similar items (e.g., documents, images, or other data points) based on their vector representations. When integrated into a retrieval system, FAISS can quickly search through large collections of vectors to return the top-k nearest neighbors for a given query vector. This makes it an excellent choice for applications such as recommendation systems, image search, and natural language processing tasks where similarity search is crucial.

```python
retriver = db.as_retriever()
docs = retriver.invoke(query)
docs[0].page_content
```

### Similartiy Search with Scores:
FAISS allows users to perform similarity searches and retrieve not only the nearest neighbors but also their corresponding similarity scores. This is particularly useful in scenarios where understanding the degree of similarity is important, such as ranking search results or filtering recommendations based on relevance.


```python
docs_and_score = db.similarity_search_with_score(query=query)
docs_and_score
```

```python
embedding_vector = embedding.embed_query(query)
embedding_vector
```

```python
docs_score = db.similarity_search_by_vector(embedding_vector)
docs_score
```
